```python
data = pd.read_csv("house_price.csv")
```

```python
data.isnull().sum()
```

```python
feature_with_na = [feature for feature in data.columns if data[feature].isnull().sum()>0]
feature_with_na
```

### Total count of feature with null values 

```python
print(f"Total count of feature with null values {len(feature_with_na)}")
```

### Find the percentage of missing values in column

```python
for feature in feature_with_na:
    col_na_percentage = round(data[feature].isnull().mean()*100,5)
    print(f"Feature :{feature}  \t| NA Value Percentage :{col_na_percentage}%")
```

### Box plot

```python
sns.boxplot(data.SalePrice)
```

```python
df = data.copy()
for feature in feature_with_na:
    data[feature] = np.where(df[feature].isnull(),1,0)
    
    df.groupby(feature)['SalePrice'].median().plot.bar()
    
    plt.title(feature)
    plt.show()
```

### Temporal Variable
Temporal variables, also known as time-based variables, are variables that represent time or are associated with time. These variables capture information about dates, times, durations, or intervals. Temporal variables are often analyzed to identify trends, patterns, seasonality, and anomalies over time.

```python
year_feature = [feature for feature in data.columns if 'Yr' in feature or 'Year' in feature]
year_feature
```

```python
for feature in year_feature:
    print(feature, data[feature].unique())
    
```

```python
df = data.copy()
df.groupby('YrSold')['SalePrice'].median().plot()
plt.xlabel('Year Sold')
plt.ylabel("Sale Price")
plt.title('House Price Vs Year Sold')
plt.show()
```

```python
df = data.copy()
df.groupby('YearBuilt')['SalePrice'].median().plot()
plt.xlabel('Year Built')
plt.ylabel("Sale Price")
plt.title('House Price Vs Year Built')
plt.show()
```

```python
df = data.copy()
df.groupby('GarageYrBlt')['SalePrice'].median().plot()
plt.xlabel('GarageYrBlt')
plt.ylabel("Sale Price")
plt.title('House Price Vs GarageYrBlt')
plt.show()
```

```python
df = data.copy()
df.groupby('YearRemodAdd')['SalePrice'].median().plot()
plt.xlabel('YearRemodAdd')
plt.ylabel("Sale Price")
plt.title('House Price VsYearRemodAdd')
plt.show()
```

### Scatter Plot

```python
df = data.copy()


for feature in year_feature:
    if feature != 'YrSold':
        df[feature] = df['YrSold']-df[feature]
        
        plt.scatter(df[feature], df['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('Sale Price')
        plt.show()
```
