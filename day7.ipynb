### extract records based on mathematical conditions

```python
query = "select * from countrylanguage where percentage >75"
data = pd.read_sql_query(query,myconn)
data
```

```python
query = "select * from countrylanguage where percentage >75 and IsOfficial='T'"
data = pd.read_sql_query(query,myconn)
data
```

```python
query = "select distinct(CountryCode) from countrylanguage where percentage >75 or IsOfficial='T'"
data = pd.read_sql_query(query,myconn)
data
```
