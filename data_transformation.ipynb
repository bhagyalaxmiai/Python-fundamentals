### One-hot encoding (Dummy Encoding)

```python
# sample data
data = {'Category' : ['A','B','A','C','B','C','D','E','F']}

df = pd.DataFrame(data)

df

```

### Label Encoding

```python
from sklearn.preprocessing import OrdinalEncoder
# Sample data

data = {'Category' : ['low','medium','high','low','high','medium','normal']}
df = pd.DataFrame(data)

#define category order
categories_order = ['low','normal','medium','high']

#initialize ordinal encoder
ordinal_encoder = OrdinalEncoder(categories=[categories_order])



df['Category_encoded'] =ordinal_encoder.fit_transform(df[['Category']])


df
```

### Ordinal Encoding

```python
#sample data
data = {'Category' : ['low','medium','high','low','high','medium']}
df = pd.DataFrame(data)

ordinal_mapping ={'low':0,'medium':1,'high':2}


df['Category_mapping'] = df['Category'].map(ordinal_mapping)
df
```

```python

data = {'Category' : ['low','medium','high','low','high','medium','normal']}
df = pd.DataFrame(data)

ordinal_mapping_1 ={'low':0,'normal':1,'medium':2,'high':3 }



df['Category_mapping'] = df['Category'].map(ordinal_mapping_1)
df
```

### Frequency encoding

```python
data = {'Category':['A','B','A','C','B','C','A','A']}
df = pd.DataFrame(data)

frequency_map = df['Category'].value_counts()
frequency_map

df['Category_frequency_mapping'] = df['Category'].map(frequency_map)
df
```

```python
data = {'Category':['A','B','A','C','B','C','A','A']}
df = pd.DataFrame(data)

frequency_map = df['Category'].value_counts(normalize=True)
frequency_map

df['Category_frequency_mapping'] = df['Category'].map(frequency_map)
df
```

### Temporal encoding

```python
data = {'DateTime' :['2022-01-06 18:12:23', '2024-10-10 10:10:10', '2023-09-03 12:12:24','2021-12-12 05:05:52'],
       'Value':[10,15,60,30]}
# data = {'DateTime' :['01-06-2033 18:12:23', '10-10-2024 10:10:10', '09-03-2023 12:12:24','12-12-2021 05:05:52'],
#        'Value':[10,15,60,30]}
df = pd.DataFrame(data)
df
df['DateTime'] = pd.to_datetime(df['DateTime'])
df['year']=df['DateTime'].dt.year
df['month']=df['DateTime'].dt.month
df['day']=df['DateTime'].dt.day
df['hour']=df['DateTime'].dt.hour
df['minute']=df['DateTime'].dt.minute
df['second']=df['DateTime'].dt.second
df
```
